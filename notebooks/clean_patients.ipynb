Checking whether all the expired patients DOD(DATE OF DEATH) is recorded

```python
mask = (patients_df['EXPIRE_FLAG'] == 1) & (patients_df['DOD'].isnull())
print(mask.sum())
```

```python
cols_to_keep = ['SUBJECT_ID','GENDER','DOB','DOD','EXPIRE_FLAG']
patients = patients_df[cols_to_keep].copy()
```

```python
patients.info()
```

```python
patients['DOB'] = pd.to_datetime(patients['DOB'], errors="coerce")
patients['DOD'] = pd.to_datetime(patients['DOD'], errors="coerce")
```

```python
check = (patients['DOD'].notnull())
print(check.sum())
```
